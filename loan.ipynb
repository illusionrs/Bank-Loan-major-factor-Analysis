# Bivariate Anaylysis

```python
#We have to understand that for which purpose and how much amount company gave loan and for which purpose mostly
#people take loan

plt.figure(figsize=(20,10))
sns.barplot(data =data, x='purpose', y='loan_amnt', hue ='loan_status')
plt.title('Purpose of Loan vs Loan Amount')
plt.show()
```

```python
#This graph will show the employee working length vs loan amount and the loan status that people paid their loan
#or in charged off.

plt.figure(figsize=(20,10))
sns.barplot(data =data, x='emp_length', y='loan_amnt', hue ='loan_status')
plt.title('EMp Length vs Loan Amount')
plt.show()
```
